```python
import mysql.connector
import pandas as pd
from prophet import Prophet

def get_value_info(year, area, sex, age):
    # MySQLに接続
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="password"
    )
    
    # カーソルを取得
    cursor = conn.cursor()
    cursor.execute("USE kobe")
    
    # SQL文をパラメータ化して実行
    sql = f"SELECT value FROM eurf310005_{year} WHERE area = %s AND sex = %s AND age = %s"
    
    # SQLクエリを実行し、結果を取得
    cursor.execute(sql, (area, sex, age))
    value = cursor.fetchone()  # 1つの結果を取得
    
    # 結果を消費する
    cursor.fetchall()  # もし何か結果が残っていたら消費する

    # カーソルと接続を閉じる
    cursor.close()
    conn.close()

    return value

# 地域、年齢リストを取得
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="password"
)

cursor = conn.cursor()
cursor.execute("USE kobe")

# 地域リストの取得
select_area_query = "SELECT DISTINCT area FROM eurf310005_2013"
cursor.execute(select_area_query)
area_list = [area[0] for area in cursor.fetchall()]

# 年齢リストの取得
select_age_query = "SELECT DISTINCT age FROM eurf310005_2013"
cursor.execute(select_age_query)
age_list = [age[0] for age in cursor.fetchall()]

# 接続を閉じる
cursor.close()
conn.close()

sex_list = ["男", "女"]
year_list = range(2013, 2024)

# 予測結果を格納するDataFrameを作成
results = []

for area in area_list:
    for sex in sex_list:
        for age in age_list:
            value_list = []
            for year in year_list:
                value = get_value_info(year, area, sex, age)
                if value:
                    value_list.append(value[0])  # 取得した値をリストに追加
            
            # DataFrameを作成
            if len(value_list) > 0:
                df = pd.DataFrame({'ds': pd.date_range(start=f"{year_list[0]}-01-01", periods=len(value_list), freq='Y'),
                                   'y': value_list})
                
                # Prophetモデルを作成
                model = Prophet()
                model.fit(df)
                
                # 未来のデータを予測
                future = model.make_future_dataframe(periods=2, freq='Y')  # 2024年と2030年の予測
                forecast = model.predict(future)
                
                # 予測結果を保存
                results.append({
                    'area': area,
                    'sex': sex,
                    'age': age,
                    '2024_value': forecast['yhat'].iloc[-2],  # 2024年の予測値
                    '2030_value': forecast['yhat'].iloc[-1]   # 2030年の予測値
                })

# 最終的な結果をDataFrameとして表示
results_df = pd.DataFrame(results)
print(results_df)
```

```python
# 値を整数に変換
results_df['2024_value'] = results_df['2024_value'].astype(int)
results_df['2030_value'] = results_df['2030_value'].astype(int)

print(results_df)
```

```python
import mysql.connector
import pandas as pd

# データベース接続設定
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': 'password',
    'database': 'kobe'
}

# データを挿入する関数
def insert_data(data):
    try:
        # データベースに接続
        connection = mysql.connector.connect(**db_config)
        cursor = connection.cursor()

        # 2024_futureテーブルにデータを挿入
        insert_query_2024 = "INSERT INTO 2024_future (area, sex, age, value) VALUES (%s, %s, %s, %s)"
        cursor.executemany(insert_query_2024, data)

        # 2030_futureテーブルにデータを挿入
        insert_query_2030 = "INSERT INTO 2030_future (area, sex, age, value) VALUES (%s, %s, %s, %s)"
        cursor.executemany(insert_query_2030, data)

        # コミットして変更を保存
        connection.commit()

        print(f"{cursor.rowcount} 行が挿入されました。")

    except mysql.connector.Error as err:
        print(f"エラーが発生しました: {err}")
    finally:
        # リソースを解放
        if cursor:
            cursor.close()
        if connection:
            connection.close()

# DataFrameをタプルのリストに変換
data_2024 = list(zip(results_df['area'], results_df['sex'], results_df['age'], results_df['2024_value']))
data_2030 = list(zip(results_df['area'], results_df['sex'], results_df['age'], results_df['2030_value']))

# データを挿入する
insert_data(data_2024)
insert_data(data_2030)
```
